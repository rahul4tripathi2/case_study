# Importing required libaries

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import LinearSVC
```

# Displaying dimensions and basic statistics of train and test data

```python
# loading given train and test dataset
train_df = pd.read_csv(r'C:\AI\case_study\multiclass_classification\train.csv')
test_df = pd.read_csv(r'C:\AI\case_study\multiclass_classification\test.csv')
```

```python
# displaying shape/dimension of train and test data
print(train_df.shape, test_df.shape)
train_df.head()
```

```python
# displaying basic statistics of train data
train_df.describe()
```

```python
# displaying basic statistics of test data
test_df.describe()
```

```python
train_df.drop(train_df.columns[0], axis = 1, inplace = True) # dropping id column since we only need features of dataset
```

```python
features = [f for f in list(train_df) if "feature" in f] # storing all the features in features variable using list comprehensions
X = train_df[features] # passing features of traindata and storing in X variable
y = train_df["label"] # storing label columns in y variable
train_df.head() # displaying train dataframe
```

```python
test_df.drop(test_df.columns[0], axis = 1, inplace = True) # dropping id column since we only need features of dataset
```

```python
features = [f for f in list(test_df) if "feature" in f] # storing all the features in features variable using list comprehensions
X_test = test_df[features] # passing features of testdata and storing in X_test variable
X_test.head() # displaying test dataframe
```

# fitting the model and validating score (spotchecking)

Random Forest Classifier

```python
rand_clf = RandomForestClassifier(n_jobs = 10) # creating random forest with 10 decision tree
rand_clf.fit(X,y) # fitting the random forest on train data
print "Random forest accuracy_score:", rand_clf.score(X, y) # printing the score of RandomForestClassifier
```

Gradient Boosting Classifier

```python
gb_clf= GradientBoostingClassifier(random_state=0).fit(X, y)  # fitting the GradientBoosting on train data
print "GradientBoostingClassifier accuracy_score:", gb_clf.score(X, y) # printing the score of GradientBoosting Classifier
```

Linear SVC(OneVsOneClassifier)

```python
one_vs_one_svc = OneVsOneClassifier(LinearSVC()) # passing LinearSVC to OneVsOneClassifier as an input
one_vs_one_svc.fit(X, y) # fitting the  OneVsOneClassifier on train data
print "LinearSVC(OneVsOne) accuracy_score:", one_vs_one_svc.score(X, y) # printing the score of GradientBoosting Classifier
```

# saving and loading the model

```python
# save the random_forest_model to disk
filename = r'C:\AI\case_study\multiclass_classification\random_forest_classifier.sav'
pickle.dump(rand_clf, open(filename, 'wb'))

# load the random_forest_model from disk
loaded_model = pickle.load(open(filename, 'rb'))
rand_score = loaded_model.score(X, y)
print(rand_score)
```

```python
# save the gb_model to disk
filename = r'C:\AI\case_study\multiclass_classification\gb_classifier.sav'
pickle.dump(gb_clf, open(filename, 'wb'))

# load the gb_model from disk
loaded_model = pickle.load(open(filename, 'rb'))
gb_score = loaded_model.score(X, y)
print(gb_score)
```

```python
# save the svc_model to disk
filename = r'C:\AI\case_study\multiclass_classification\one_vs_one_classifier.sav'
pickle.dump(one_vs_one_svc, open(filename, 'wb'))

# load the svc_model from disk
loaded_model = pickle.load(open(filename, 'rb'))
svc_score = loaded_model.score(X, y)
print(svc_score)
```

# Accuracy Report

```python
accuracy_report = {'Model': ['Random Forest', 'GradientBoosting', 'OneVsOneClassifier(SVC)'],
         'model_accuracy': [rand_score, gb_score, svc_score]}
df = pd.DataFrame.from_dict(accuracy_report)
df
```
